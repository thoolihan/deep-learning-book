# Using OpenCV for facial detection
Example from this [article](https://towardsdatascience.com/face-detection-in-2-minutes-using-opencv-python-90f89d7c0f81)

```python
import cv2
```

```python
import os
HOME = os.path.expanduser("~")
CLASSIFIER_PATH = os.path.join(HOME, "workspace", "opencv", "haarcascade_frontalface_default.xml")
face_cascade = cv2.CascadeClassifier(CLASSIFIER_PATH)
print(CLASSIFIER_PATH)
```

```python
%matplotlib inline

from matplotlib import pyplot as plt

def show_cv2(img, title='picture'):
    tmp = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(tmp)
    plt.title(title)
    plt.show()
```

```python
face_cascade = cv2.CascadeClassifier(CLASSIFIER_PATH)
```

```python
IMG_NAME = "family.jpg"
IMG_PATH = os.path.join("data", "faces", IMG_NAME)
img = cv2.imread(IMG_PATH)
show_cv2(img, title=IMG_NAME)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
show_cv2(gray, "(Gray) {}".format(IMG_NAME))
```

```python
# detect faces
faces = face_cascade.detectMultiScale(gray, 1.1, 4)
faces
```

```python
RECT_COLOR = (250, 0, 0)

for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), RECT_COLOR, 3)
```

```python
show_cv2(img, title="(with faces) {}".format(IMG_NAME))
```
