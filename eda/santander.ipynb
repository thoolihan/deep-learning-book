```python
import pandas as pd


INDEX_COL="ID"
LABEL = 'target'
PROJECT_NAME="santander"
INPUT_DIR="data/{}".format(PROJECT_NAME)
train_df = pd.read_csv("../{}/train.csv".format(INPUT_DIR))
```

```python
train_df.head()
```

```python
cols = list(train_df)
features = [col for col in cols if col not in [INDEX_COL, LABEL]]
```

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)

for i in range(5):
    print("Summary for '{}' column".format(features[i]))
    print(train_df[features[i]].describe())
    print("\n")
```
